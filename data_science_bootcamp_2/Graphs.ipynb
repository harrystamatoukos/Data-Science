# Combine matplotlib and Pandas plotting 

```python
ax = df.plot(kind='bar', figsize=(12,6))

# Set a title
ax.set_title('This is a plot', fontsize = 20)

# xaxis labels
ax.set_xlabel('this x axis means nothing')

#yaxis lables
ax.set_ylabel('This is the y axis')

# change the location of the legend
ax.legend(loc = 2, bbox_to_anchor=(1.02, 1))
```

Location String	Location Code
- 'best'	0
- 'upper right'	1
- 'upper left'	2
- 'lower left'	3
- 'lower right'	4
- 'right'	5
- 'center left'	6
- 'center right'	7
- 'lower center'	8
- 'upper center'	9
- 'center'	10
