# Section 5: Spit-Apply-Combine

```python
#Split: We group the data by a column
#Apply: For each group we apply a function
#Combine: We combine the split and the apply together 
```

```python
drinks_columns = ['country','beer','spirit','wine','liters','continent']

#Replace the columns during reading a file
drinks = pd.read_csv('data/drinks.csv',header = 0 , names= drinks_columns)
```

```python
drinks
```

```python
drinks.groupby('continent').beer.mean().plot(kind='bar')
```

```python
drinks.groupby('continent').mean()
```

```python
drinks.groupby('continent').beer.describe()
```

```python
drinks.groupby('continent').beer.agg(['count','mean','max']).sort_values('mean', ascending = False)
```

```python
drinks.groupby('continent').agg(['count','mean','max'])
```

```python
drinks.groupby('continent').describe()
```

```python
drinks.groupby('continent').describe()
```
