```python
cols = ['country','beer','spirit','wine','liters','continent']
drinks = pd.read_csv('data/drinks.csv' , names = cols, header = 0)
```

```python
drinks
```

```python
#Calculate the median
median = drinks.liters.median()
```

```python
drinks['volume'] = drinks.liters.apply(lambda x: 'large' if x > median else 'low')
```

```python
drinks
```

```python
drinks['volume_2'] = drinks.liters.apply(lambda x: 'very large' if x > 10 else 'large' if x > median else 'low')
```

```python
drinks
```

```python
drinks['mL'] = drinks.liters * 1000000
```

```python
drinks = drinks[['country','beer','spirit','wine','liters','mL','continent','volume','volume_2']]
```

```python
drinks
```

```python
drinks.dtypes
```

```python
drinks['mL'] = drinks.mL.apply(lambda x: int(x))
```

```python
drinks.dtypes
```

```python
def round_whole(x):
    
    rounder = round(x, 0)
    return rounder

```

```python
round_whole(6.49)
```

```python
drinks['liters_rounded'] = drinks.liters.apply(lambda x: round_whole(x))
```

```python
drinks.liters
```

```python
drinks
```
